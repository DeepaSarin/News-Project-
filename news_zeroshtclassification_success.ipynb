```python
dict1={}
classes = ['coal','petroleum','agriculture or food','maritime','steel and iron','construction','fertilizer']
sentence = "Shipping is a noble resource"
dict1=classifier(sentence, classes)
print(dict1['labels'],dict1['scores'])
```
